IBorrow Desk is a tool for monitoring borrow rates and availability using Interactive Broker's freely available data.

I worked in finance for the past five years, with my last year or so doing a decent amount of short selling. One of the resources I used quite a lot was Interactive Broker's Stock Loan Availability Database especially for hard to borrow stocks.

But it was a pain to keep updated and an overall a pain to use. So, after taking Udacity's Full Stack Nanodegree I figured this database would be a fun place to start building a working (maybe even useful) web app.

Where does IBorrowDesk get its data?
All IBorrowDesk data is sourced from Interactive Brokers. You can check the source .csv files here.

It's important to understand that because IBorrowDesk only reports data from a single (albeit very large) brokerage, the fee and availability reported may be different at other brokerages.

A stock I am following at IBorrowDesk has stopped reporting data. Is something broken?
Unfortunately, if Interactive Brokers does not have any shares of a security available to borrow they will simply stop reporting any data at all instead of reporting "0". This can happen from time to time for hard to borrow securities. Usually someone returns some shares to Interactive Brokers, or a client makes additional shares available to borrow ensuring the security can be tracked again

What does "fee" rate mean?
A stock loan fee is the fee charged by a brokerage for borrowing shares. In IBorrowDesk's case, the fee rate is expressed as an annualized percentage charged by Interactive Brokers.

More info: https://www.investopedia.com/terms/s/stock-loan-fee.asp

What does "available" mean?
In order to borrow shares from a brokerage (IBorrowDesk monitors Interactive Brokers), that brokerage needs shares available to lend out. Typically these belong to other clients of the firm, or they could belong to Interactive Brokers itself. Liquid large cap securities are typically "easy" to borrow. That is, brokerages have plenty of clients willing to lend out their shares. Some securities are in high demand - many people want to borrow them, or some may have restricted supply - few people willing to lend them. In either case the "available" number can be low

Matt levine (Bloomberg News) discusses an interesting case with Nikola where very few shares were available to borrow due to convertible arbitrage demand: People Loved Nikola But Not Its Warrants

More info: https://www.investopedia.com/terms/s/securitieslending.asp

A stock I am following doesn't have many shares available to borrow. Is that good or bad?
Unfortunately all this means is there are a lot of market participants trying to borrow shares - and not many market participants are willing to lend shares. Sometimes this can mean that a lot of short sellers take a dim view of the prospects of the company. Other times it can due to some technical restriction on trading: for example, a large issue of a publicly traded convertible bond that many bond-holders are trying to "hedge" by selling the security short.

It's also worth noting that IBorrowDesk only tracks Interactive Brokers data - it could be that other brokerages have many shares available. Professional or institutional investors have access to data sources far more sophisticated than this site - which can only track a single, albeit very large and active, brokerage.

More info: https://www.investopedia.com/terms/h/hardtoborrowlist.asp
