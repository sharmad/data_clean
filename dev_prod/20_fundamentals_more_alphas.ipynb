### 1. Advanced Structural Float & Liquidity (SF1 & SF3)

Beyond just the point-in-time basic float approximation we discussed, you can map the absolute velocity of a squeeze.

* **The Alpha:** Track institutional "dead float." The SF3 dataset maps 13F institutional ownership pointwise. For low-float gappers, if you take the basic shares outstanding, subtract insider holdings (SF2), and further subtract long-term passive index institutional holdings (like Vanguard and BlackRock from SF3), you get the true **tradable market float**.
* **Intraday Play:** If a stock gaps pre-market on an earnings catalyst and you calculate that the true tradable float is only 1.5 million shares (due to deep institutional lockups from SF3), and the stock prints 3 million shares of volume in the first 15 minutes of the market open, you have a **Float Rotation Factor of 2.0x**. When a low-float stock rotates its entire float multiple times intraday, normal resistance levels break, short-sellers get trapped, and the setup transforms into an exponential multi-day runner.

### 2. Form 4 Insider Cluster Buying & Selling (SF2)

Most swing traders use retail scanner sites that post insider buying with a 3-day lag. The SF2 dataset lets you backtest precise behavioral patterns.

* **The Alpha:** Ignore routine insider selling (which is usually automated 10b5-1 plans for tax liquidity). Instead, filter for **Open-Market Cluster Purchases**. When multiple C-suite executives (CEO, CFO, and Directors) buy shares with their own cash in the open market within a 7-day window, it creates a massive structural floor for the stock.
* **Swing Play:** Combine a technical mean-reversion setup (e.g., a heavily shorted stock hitting an oversold RSI or a major weekly moving average) with a recent point-in-time Insider Cluster Buy signal. This filter filters out falling knives and isolates stocks where management is actively preparing a turnaround or a positive PR catalyst, drastically improving your swing win-rate.

### 3. Institutional "Herding" and Smart Money Inflows (SF3)

While 13F filings are delayed by up to 45 days after quarter-end, they act as an incredible regime filter for swing setups.

* **The Alpha:** Instead of looking at who *owns* a stock, track the *delta*—specifically **institutional concentration**. Count the number of unique institutions entering a micro-cap stock for the first time.
* **Swing Play:** Micro-cap stocks that experience a sudden spike in institutional count (even small positions) are transitioning from "toxic pump-and-dump" territory to "legitimate institutional interest." If you overlay this regime filter on your momentum strategies, you can screen out bankruptcies/scams and target higher-quality small-cap names that have the institutional backing to sustain an upward multi-week trend instead of crashing back down in 24 hours.

### 4. Cross-Asset Tailwinds via ETF Flow Infiltration (SFP)

The **Sharadar Fund Prices (SFP)** database tracks daily prices and historical distributions for thousands of US ETFs and mutual funds.

* **The Alpha:** Small and micro-cap stocks are heavily prone to mechanical buying or selling caused entirely by passive index fund inflows and outflows. If a specific thematic ETF (like a clean-energy, biotech, or small-cap momentum ETF) is receiving massive capital inflows, the managers are *forced* to buy underlying component stocks basket-style.
* **Intraday/Swing Play:** Map out the smaller constituents of highly volatile thematic ETFs. When the main ETF breaks out on high volume, look for the lowest-float, most illiquid stock inside that ETF basket. Because the ETF manager must deploy cash proportionally, a sudden inflow into the ETF creates an artificial, highly predictable buying squeeze on its lowest-float constituent, giving you a massive systematic advantage over traders only watching the individual stock's chart.

## SF3 / SF3A / SF3B — 13F Institutional Holdings

This is the most underused dataset in the bundle for short-term traders. The standard use is tracking aggregate institutional ownership. The alpha is elsewhere:

**Float-adjusted institutional crowding.** Once you have your PIT float estimate from SF1 + SF2, divide institutional shares held (SF3A) by float rather than shares outstanding. That's the number that actually matters. When float-adjusted institutional ownership is 60–80%+, you've identified a stock where any catalyst forces institutions to compete for a thin float. That's the mechanical precondition for a genuine squeeze or momentum extension — not just "low float" in isolation.

**Institution quality tiering.** SF3B lets you identify which specific institutions hold what. The signal from a Millennium or Citadel 13F is completely different from a regional RIA accumulating shares. Build a simple quality score for each institution based on their historical hit rate on positions — funds that systematically appear in stocks 1–2 quarters before a large move. Track their new positions as a leading indicator for your watchlist generation. This is effectively free factor research.

**Institutional vacuum on short setups.** For mean-reversion short setups on gap-and-go names that have run irrationally, high and *declining* institutional ownership (multi-quarter trend) combined with rising insider selling from SF2 is a structurally bearish setup before the price even shows it.

**New position emergence.** A stock with < 5 institutional holders suddenly gaining 3 new 13F filers in one quarter, especially if those filers have alpha track records, is a pre-breakout signal for swing setups.

## SF2 — Insiders Beyond Float

You're already using this for float calculation. The additional alpha signals:

**Cluster buying is the highest-quality insider signal that exists.** Three or more insiders making open-market purchases within a 30-day window is statistically extremely strong, particularly for sub-$500M market cap companies where insiders have genuine information advantage. Academic literature (Jeng, Metrick, Zeckhauser) quantifies this at 6–10% annualised alpha over 6-month horizons. For your swing timeframe, it's a watchlist trigger: when cluster buying appears, wait for a technical catalyst (volume surge, range breakout) and take the trade with a fundamental tailwind.

**10b5-1 plan vs. non-plan purchases.** The `transactioncode` field lets you separate these. Non-plan open-market purchases (`P` code) are genuinely discretionary and therefore informative. Automatic plan sales (`S` code from a 10b5-1) carry no information — insiders set those up months in advance specifically to avoid the appearance of informed trading. Most retail tools conflate these; you won't.

**Sequential purchasing.** The same insider making multiple open-market purchases across successive weeks or months is a much stronger signal than a single purchase. It suggests conviction is increasing, not just a one-off. Easy to detect by grouping SF2 rows by `(ticker, insidertitle)` and checking for consecutive transaction dates within a rolling 60-day window.

**Purchase size relative to existing position.** `transactionshares / sharesownedfollowingtransaction` — if a CEO buys 10,000 shares but already owns 5M, it's noise. If they double their position, it's signal. Trivial to compute from SF2, and almost no retail tooling does it.

## SF1 — Fundamentals for Strategy-Specific Use Cases

For swing and intraday strategies on small/micro-cap names, the useful signals are not the classic value factors (P/E, P/B) — those work on 6–18 month horizons. What's relevant:

**Accruals anomaly in small caps.** Net income minus operating cash flow, scaled by assets. High accruals (earnings driven by accounting rather than cash) predicts underperformance. This is well-documented and specifically stronger in small caps where analyst coverage is thin. For your short setups on over-extended gap-and-go names, a high accruals ratio is structural confirmation the move isn't supported by real economics.

**Asset growth anomaly.** Companies aggressively growing assets (acquisitions, capex) underperform. Again strongest in small caps. Simple to compute as year-over-year change in total assets. Useful for filtering your universe — avoid going long on names where aggressive asset growth suggests empire-building management.

**Working capital deterioration as distress signal.** Declining current ratio + increasing days payable outstanding + declining cash is a leading indicator of financial distress 2–4 quarters ahead. For short setups specifically, combining this with a gap-up catalyst (where retail is excited) gives you a structurally weak name to fade.

**Revenue growth momentum.** Sequential quarter-over-quarter revenue acceleration (not just year-over-year beat) is a stronger predictor of near-term price momentum than earnings beats alone. For EP (episodic pivot) gap-up setups specifically, the quality of the gap depends partly on whether the revenue inflection is genuine and accelerating. Companies with decelerating revenue that beat on cost-cuts are lower-quality setups.

## DAILY — Computed Daily Metrics

Often overlooked because it looks like a convenience dataset, but there are two specific uses:

**Daily market cap for dynamic universe filtering.** Your low-float screens need to be applied to the correct market-cap band each day — not just at point of portfolio construction. DAILY gives you market cap for every stock every day, which combined with your SF1-derived float estimate gives you a daily float-adjusted market-cap sort. This matters because a stock can enter and exit your target universe across a 10-year backtest as it grows or gets diluted.

**EV/EBITDA for quality gating.** For gap-up swing setups specifically, filtering out negative-EBITDA names (or extremely high EV/EBITDA) removes the regime where a company is pre-revenue/pre-profit and therefore technically "anything" can justify the gap. Your hit rate on EP setups will improve if you gate on companies with at least some earnings foundation.

## SP500 Historical Constituents

This one is genuinely underused and has two concrete applications:

**Index addition/deletion events.** S&P 500 additions are announced with typically 5–7 days' notice before effective date. Passive funds must buy at the close on effective date. Active funds front-run. The stock reliably pops on announcement and then fades after effective date (the "index addition effect" is well documented). For your swing setup pipeline this is a *dated catalyst* with known mechanics — exactly the kind of setup that systematises cleanly.

**Russell reconstitution.** More relevant to your universe than S&P 500 because Russell rebalances across the entire small-cap and micro-cap spectrum in June. The annual reconstitution creates predictable buying pressure on index additions (stocks crossing the Russell 2000 market-cap threshold) and selling pressure on deletions. For low-float small caps this effect is amplified — the same buying demand hits a thinner float. This is entirely derivable from the SP500 dataset (for S&P membership history) combined with DAILY market cap to infer Russell eligibility. Russell doesn't publish a clean historical membership file at this price point, but you can reconstruct it reasonably well from market-cap rankings.
