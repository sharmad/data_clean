If your code uses raw Close prices instead of Adjusted Close prices, your moving averages will trigger fake breakdown signals every time the ETF goes ex-dividend.

## Practical Notes

**Spread tax reality by tier.** For a $50k account rebalancing frequently: [A]/[B] ETFs have sub-$0.01 spreads and you can ignore impact. [C]/[D] ETFs run $0.01–$0.03 and are fine unless you're rebalancing daily. [E] ETFs hit $0.03–$0.07 — you're paying 2–5bps per side, still manageable if turnover is weekly/monthly. [F] ETFs (EVX, CRAK, PBS) can gap to $0.10+ intraday — time your entries to liquid market hours (10:00–15:30 ET) and use limit orders.

**Biotech dual-use warning.** XBI and IBB are both "Biotechnology" but represent completely different factor exposures. XBI is equal-weighted small/mid biotech with binary FDA-event risk and 50%+ annual drawdowns in stress. IBB is cap-weighted mega-cap biotech acting more like a defensive growth ETF. Running both simultaneously is intentional only if you want separate large/small biotech legs.

**Communication Services post-2018 composition.** XLC is highly concentrated — the top 5 names (GOOGL, META, NFLX, ATVI, DIS historically) can represent 40%+ of the fund. VOX is marginally more diversified. IYZ gives you the old telecom-only exposure (AT&T, Verizon, T-Mobile) without the internet giants.

**MLP tax structures.** AMLP uses a C-corp wrapper which creates a tax drag of roughly 1–2% annually relative to owning the MLPs directly. MLPA and MLPX avoid this. For systematic strategies with frequent rebalancing, the structure matters less than for buy-and-hold; just model it correctly in your cost assumptions.

**Semiconductor concentration.** SMH has had >20% in NVDA at peak. SOXX is more diversified by index construction (capped weighting). If you're using both in a multi-strategy portfolio, you're doubling semiconductor factor exposure — be explicit about this in your factor risk model.

**Verify before going live.** ETF tickers and AUM shift. Any fund with [E] or [F] tier should be checked on ETF.com or the issuer's site for current AUM, expense ratio, and whether the fund is still in operation before including in any live strategy.

### 🧠 The Quant's Survival Guide for Trading These (Read Carefully)

Since you are running automated/systematic strategies on a low budget, I'm going to save you thousands of dollars right now:

1.  **Beware of "Stale Pricing" (Time Zone Disconnect):** 
    When you are trading `EWJ` (Japan) or `EWG` (Germany) at 2:00 PM EST, **their underlying stock markets are closed**. The ETF price you see is being maintained by US market makers using correlated US assets (like S&P futures or currency pairs) to guess where the foreign market *will* open. If your model uses intraday standard deviation or breakout logic on foreign ETFs during US hours, your signals are mostly reacting to US market noise, not local country news. 
2.  **Trade at the European Close:**
    If your system trades European ETFs (`EWG`, `EWU`, `EWQ`), execute your trades between 10:30 AM and 11:30 AM EST. The European markets close at 11:30 AM EST. After that, liquidity dries up and spreads widen because the European market makers log off.
3.  **Spread-to-Price Ratio:**
    Because you have a small account, you might be tempted to buy cheap ETFs. Don't look at the share price; look at the **spread as a percentage of the price**. A 2-cent spread on a $10 ETF is a 0.2% loss the second you enter. A 2-cent spread on a $50 ETF is a 0.04% loss. Stick to the tickers above; they usually trade at a 1-cent spread.
4.  **Currency Drag (Unhedged vs. Hedged):**
    All the `E-series` ETFs above are **unhedged**. That means if you buy `EWG` (Germany), you are long German stocks *and* long the Euro. If the Euro drops against the Dollar, your ETF goes down even if German stocks went up. If your model is specifically trying to trade equity momentum, you might want to look at currency-hedged ETFs (like `HEWJ` for Japan or `HEWG` for Germany), though they are less liquid.
5.  **Avoid Frontier Markets:**
    You'll see ETFs for places like Nigeria, Vietnam, or Colombia on Finviz. **Ignore them.** Their bid-ask spreads are wide enough to drive a truck through. As a retail systematic trader, if an ETF doesn't trade at least 1,000,000 shares a day, exclude it from your universe.

### ⚠️ The Quant's Warning on Leveraged & Inverse ETFs

Since you are coding a systematic strategy, handling `TQQQ`, `SQQQ`, `SOXL`, or `SOXS` requires extreme caution. If you naively treat these like normal stocks in your backtester, your results will be dangerously inaccurate. Here is what you must account for in your code:

1. **The "Volatility Drag" (Beta Slippage):**
   These ETFs are designed to track the *daily* percentage change of the index, not the long-term change. 
   * *The Math:* If the QQQ goes down 10% on Monday, `TQQQ` drops 30%. To get back to even, the QQQ needs to go up 11.1%. But `TQQQ` now needs to go up **42.8%** just to break even! 
   * *The Code Implication:* Do not hold 2x/3x ETFs in a systematic swing or trend-following strategy that keeps positions open for months, especially during choppy, sideways markets. The math practically guarantees the ETF will decay over time. Use them for short-term mean-reversion (1-5 days) or pure intraday momentum.
2. **Reverse Splits Will Break Bad Data:**
   Because of volatility drag, inverse ETFs (like `SQQQ` and `SOXS`) inevitably grind toward zero over the years. To keep the share price tradeable, the fund managers frequently execute reverse splits (e.g., 1-for-5 or 1-for-10). If your data provider doesn't correctly handle historical split adjustments, your backtest will see a "jump" from $5 to $25 overnight and register a massive fake profit or loss. Always sanity-check inverse ETF data.
3. **Calculate Signals on the Underlying, NOT the Leveraged ETF:**
   If you want to trade `TQQQ` using a moving average crossover or an RSI signal, **do not calculate the indicator on TQQQ's price data**. Calculate your RSI and Moving Averages using `QQQ` (the underlying asset). Generate the trade signal from `QQQ`, and then execute the trade in `TQQQ`. Why? Because the daily compounding of the leveraged ETF alters the historical volatility profile, which warps technical indicators and statistical z-scores.
4. **Hard-to-Borrow (HTB) Costs:**
   If your strategy attempts to short inverse ETFs (e.g., shorting `SQQQ` to capture its structural decay—a common quant trade), be aware that these are often Hard-to-Borrow. Retail brokers will charge you hefty annualized borrow fees. Ensure your backtester can simulate these borrow costs, or your 15% annualized backtested gain might actually be a net loss after fees.

For a low-budget systematic trader, this ETF universe is actually powerful because it lets you model:

* Global country rotation
* Sector rotation
* Industry rotation
* Risk-on / risk-off
* Bear market leadership
* Relative strength
* Intermarket relationships

without needing expensive survivorship-bias-free constituent databases for every stock universe. In practice, many institutional top-down models are built from a universe of only ~100-300 highly liquid ETFs rather than thousands of individual equities.
